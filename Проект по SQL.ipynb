## Ответы на вопросы

### Посчитайте, сколько книг вышло после 1 января 2000 года

```python
books_2000 = '''
SELECT COUNT(book_id) FROM books WHERE publication_date > '2000-01-01'
'''
read(books_2000)
```

После 1 января 2000 года вышло 819 книг

### Посчитайте для каждой книги количество обзоров и среднюю оценку

```python
books_rating_avg = '''                              
                    SELECT
                        b.title AS book_title,
                        b.book_id AS book_id,
                        AVG(rat.rating) AS avg_rating,
                        COUNT(DISTINCT rev.review_id) AS count_reviews 

                    FROM
                        books AS b
                    LEFT JOIN reviews AS rev ON rev.book_id=b.book_id
                    LEFT JOIN ratings AS rat ON rat.book_id=b.book_id
                    

                    GROUP BY
                        b.title,
                        b.book_id

                    ORDER BY
                        count_reviews DESC;
                ''' 
read(books_rating_avg).round()
```

Максимальное кол-во обзоров у книги Twilight (Twilight #1) - 7

### Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры

```python
books_max_50 = '''
            SELECT b.publisher_id, p.publisher, COUNT(*) count_books
            FROM books AS b
            LEFT JOIN publishers AS p ON b.publisher_id = p.publisher_id
            WHERE num_pages > 50
            GROUP BY b.publisher_id, p.publisher
            ORDER BY COUNT(b.book_id) DESC
            LIMIT 1
        '''
read(books_max_50)
```

Издательство Penguin Books - выпустило наибольше число книг,толще 50 страниц в кол-ве 42 книг

### Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками

```python
autor_max_50 = '''
            SELECT a.author,
                    AVG(r.rating) AS avg_rating
            FROM authors AS a
            LEFT JOIN books b ON a.author_id = b.author_id
            LEFT JOIN ratings r ON b.book_id = r.book_id
            WHERE b.book_id IN (SELECT book_id
                                FROM ratings
                                GROUP BY book_id
                                HAVING COUNT(rating) >=50)
            GROUP BY a.author   
            ORDER BY avg_rating DESC
            LIMIT 1
            '''
read(autor_max_50)
```

Cамая высокая средняя оценка у автора J.K. Rowling/Mary GrandPré (4.287)

### Посчитайте среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
avg_user_50='''
            SELECT ROUND (AVG(count))
            FROM (SELECT COUNT(review_id)
                  FROM reviews AS r
                  WHERE r.username IN (SELECT username
                                       FROM ratings
                                       GROUP BY username
                                       HAVING COUNT(rating) > 50)
                  GROUP BY r.username) AS cnt;

        '''
read(avg_user_50)
```

Среднее кол-во обзоров пользователей, которые поставили более чем по 50 оценок равно 24.
