```python
# Добавим столбец со стоимостью квадратного метра и приведем его в целочисленный тип
flat['sm_price'] = flat['last_price'] / flat['total_area']
flat['sm_price'] = flat['sm_price'].astype(int)
```

```python
# Добавим столбцы дня недели, месяца и года
flat['weekday'] = flat['first_day_exposition'].dt.weekday
flat['month'] = flat['first_day_exposition'].dt.month
flat['year'] = flat['first_day_exposition'].dt.year
```

```python
# Добавим столбец с отношением жилой площади к общей, переведем в % и приведем его в целочисленный тип
flat['area_ratio'] = flat['living_area'] / flat['total_area'] * 100
flat['area_ratio'] = flat['area_ratio'].fillna(value=0).astype(int)
```

```python
# Добавим столбец с отношением площади кухни к общей, переведем в % и приведем его в целочисленный тип
flat['kitchen_ratio'] = flat['kitchen_area'] / flat['total_area'] * 100
flat['kitchen_ratio'] = flat['kitchen_ratio'].fillna(value=0).astype(int)
```

```python
# Создадим функцию для заполнения нового столбца с типом этажа
def floor_group(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
flat['floor_type'] = flat.apply(floor_group, axis=1)
```
